# Latihan Visualisasi Data

```python
import matplotlib.pyplot as plt

# Data contoh
bulan = ['januari', 'februari', 'maret', 'april', 'mei', 'juni', 'juli', 'agustus', 'september', 'oktober', 'november', 'desember'] 
berat_badan = [97, 90, 86, 85, 80, 79, 81, 78, 76, 72, 69, 65]

# grafiknya
plt.figure(figsize=(12, 7))
plt.plot(bulan, berat_badan, marker='*', color='r', linestyle='-') 
plt.title('Line Chart Penurunan Berat Badan per bulan')
plt.xlabel('Bulan')
plt.ylabel('Berat Bdan')
plt.grid(True)
plt.show()
```

grafik diatas merupakan visualisasi data menggunakan library matplotlib dari data penurunan berat badan per bulan, selama 12 bulan.

```python
import matplotlib.pyplot as plt

# Data contoh
hari = ['senin', 'selasa', 'rabu', 'kamis', 'jumat', 'sabtu', 'minggu']
jumlah_pengunjung = [256, 179, 188, 302, 445, 704, 535]
#grafiknya
plt.figure(figsize=(12, 7))
plt.bar(hari, jumlah_pengunjung, color='g')
plt.title('Bar Chart Pengunjung Cafe per hari')
plt.xlabel('Hari')
plt.ylabel('Jumlah Pengunjung')
plt.grid(False)
plt.show()
```

grafik histogram diatas merupakan data visualisasi menggunakan library matplotlib data dari jumlah pengunjung cafe per hari, selama 1 minggu.

```python
import matplotlib.pyplot as plt

#Data contoh
kota = ['kampar', 'duri', 'bengkalis', 'dumai', 'minas', 'rengat', 'rohil']
mahasiswa_rantau = [13, 89, 16, 60, 48, 20, 22]

#grafiknya
plt.figure(figsize=(7, 5))
plt.pie(mahasiswa_rantau, labels=kota, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart Mahasiswa Merantau')
plt.axis('equal') # Agar Lingkaran berbentuk Lingkaran sempurna
plt.show()
```

pie chart di atas merupakan visualisasi data menggunakan library matplotlib dari chart mahasiswa rantau yang berada dibeberapa kota provinsi riau.

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Data Contoh
usia = [15, 16, 17, 18, 19, 20]
uang_jajan = [200, 250, 350, 500, 650, 800]

#Scatter plot #1
plt.figure(figsize=(6,4))
sns.scatterplot(x=usia, y=uang_jajan)
plt.title('Kebutuhan Hidup Menurut Usia #1')
plt.xlabel('usia')
plt.ylabel('uang jajan')
plt.grid(True)
plt.show()

#Scatter plot #2
plt.figure(figsize=(6, 4))
sns.regplot(x=usia, y=uang_jajan)
plt.title('Kebutuhan Hidup Menurut Usia #2')
plt.xlabel('usia')
plt.ylabel('uang jajan')
plt.grid(True)
plt.show()
```

Merupakan visualisasi data menggunakan library seaborn dari data uang jajan dengan usia.
